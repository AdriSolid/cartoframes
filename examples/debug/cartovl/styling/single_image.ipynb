```python
import cartoframes
from cartoframes.carto_vl import carto
from cartoframes import CartoContext

context = CartoContext(
    base_url='https://cartovl.carto.com/', 
    api_key='default_public'
)

carto.Map(
    [carto.Layer(
        'populated_places',
        symbol_='star'
    )],
    context=context
).init()
```
