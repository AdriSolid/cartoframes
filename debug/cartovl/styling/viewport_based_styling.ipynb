```python
import cartoframes
from cartoframes.contrib import vector
from cartoframes import CartoContext

context = CartoContext(
    base_url='https://cartovl.carto.com/', 
    api_key='default_public'
)

vector.vmap(
    [vector.Layer(
        'spend_data',
        color_='ramp(linear($amount, viewportMax($amount), viewportMin($amount)), Bluyl)',
        width_='sqrt($amount) / 4',
        stroke_color_='rgba(51, 51, 51, 0.3)'
    )], 
    context=context
)
```
